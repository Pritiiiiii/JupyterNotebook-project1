```python
import pandas as pd

# Load the Excel file
file_path = r"D:\data\Tyba.xlsx"  # Replace with your file path
df = pd.read_excel(file_path)

# Display a sample of the data for debugging
print("Original Data Sample:")
print(df.head())

# Ensure the PINCODE column is treated as integers
if df['PINCODE'].dtype != 'int64':  # Convert to integers if needed
    df['PINCODE'] = pd.to_numeric(df['PINCODE'], errors='coerce', downcast='integer')

# Define the list of PIN codes based on your input
pincode_list = [
    400058, 400053, 400069, 400068, 400094, 400060, 400102, 400104,
    400097, 400064, 400063, 400087, 400065, 400067, 400092, 401107,
    401101, 401201, 401202, 401203
]

# Debug: Display the PINCODE list
print("Pincode List:")
print(pincode_list)

# Filter the data
filtered_df = df[df['PINCODE'].isin(pincode_list)]

# Debug: Display the filtered data
print("\nFiltered Data Sample:")
print(filtered_df.head())

# Check if filtered data is empty
if filtered_df.empty:
    print("No matching data found for the provided pincodes.")
else:
    # Save the filtered data to a new Excel file
    output_file = r"D:\data\filtered_data.xlsx"
    filtered_df.to_excel(output_file, index=False, engine='openpyxl')  # Specify the engine explicitly
    print(f"\nFiltered data saved to: {output_file}")
```
